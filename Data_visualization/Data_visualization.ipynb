Dataset used for the problem is **'State/UT-wise Accidents Classified According to Type of Traffic Violations during 2018'** from 'https://data.gov.in/resource/stateut-wise-accidents-classified-according-type-traffic-violations-during-2018'



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Road-Accidents-2018-Annexure-36 - Road-Accidents-2018-Annexure-36.csv",header=0)
df = df.drop(df.index[-1])
```

**Bar plot and line plot**

```python
states = df['States/UTs'].tolist()
OS_num = df['Over-Speeding - Number of - Number'].tolist()
OS_kills = df['Over-Speeding - Persons Killed - Number'].tolist()
OS_injured = df['Over-Speeding - Persons Injured - Total Injured'].tolist()
plt.figure(figsize=(10,4),dpi=100)
plt.bar(states,OS_num,label='Overspeed No. of accidents')
plt.plot(states,OS_kills,'-or',label= 'Overspeed deaths')
plt.plot(states,OS_injured,'-oc', label='Overspeed injured')
plt.xticks(rotation = 90)
plt.legend()
plt.ylabel('Number')
plt.title('State-wise number of overspeed reports')
plt.grid(axis='y')
plt.show()
```

By analyzing the data, the numbers for the cases due to overspeeding are very high compared to others. So it is good to plot the same separately. From the above plot, it can be inferred that the number of accidents due to overspeeding is high in 6 states, crossing 20,000 cases. The number of injuries due to overspeeding is more than the number of cases in some states which shows that more than a single person was injured in some accidents, it can be co-passengers or third parties. In some states such as Bihar, and Punjab, the number of deaths to the total cases ratio is higher. The same is evident from the plot as the number of death pointers lies in the upper part of the bar representing a number of cases. 

**Line plot**

```python

DD_kills = df['Drunken Driving/Consumption of Alcohol & Drug - Persons Killed'].tolist()
DWS_kills = df['Driving on Wrong Side - Persons Killed'].tolist()
JRL_kills = df['Jumping Red Light - Persons Killed'].tolist()
UM_kills = df['Use of Mobile Phone - Persons Killed'].tolist()
#plt.plot(states,OS_kills)
plt.figure(figsize=(10,4),dpi=100)
plt.plot(states,DD_kills,label='Drunken Driving/Consumption of Alcohol & Drug')
plt.plot(states,DWS_kills,label='Driving on Wrong Side')
plt.plot(states,JRL_kills,label='Jumping Red Light')
plt.plot(states,UM_kills,label='Use of Mobile Phone')
plt.xticks(rotation = 90)
plt.legend(title='Cause')
plt.ylabel('Number')
plt.grid(axis='x')
plt.title('State-wise number of deaths by accident')
plt.show()
```

In this plot, the number of deaths by accidents due to causes other than overspeeding is analyzed. The accidents caused by 'Jumping red light' have less death toll compared to others. The statistics of UP are alarming as the deaths due to three violations, 'drunken driving/consumption of alcohol & drug', 'driving on wrong side' and 'use of mobile phone' is very high even though overspeeding cases were less as seen previously. 'Driving on the wrong side' is a major cause of death in many states. This must be due to head-on collision due to traffic coming from the opposite direction which can be fatal.

**Scatter Plot**

```python
df['DD_rank'] = df['Drunken Driving/Consumption of Alcohol & Drug - Number of Accidents'].rank(method='min')
df['OS_rank'] = df['Over-Speeding - Number of - Number'].rank(method='min')
df['DW_rank'] = df['Driving on Wrong Side - Number of Accidents'].rank(method='min')
df['JRL_rank'] = df['Jumping Red Light - Number of Accidents'].rank(method='min')
df['UM_rank'] = df['Use of Mobile Phone - Number of Accidents'].rank(method='min')
data_sort_rank = df.sort_values(by=['OS_rank','DD_rank','DW_rank','JRL_rank','UM_rank'])
plt.figure(figsize=(10,4),dpi=100)
plt.scatter(data_sort_rank['States/UTs'],data_sort_rank['OS_rank'],label='Overspeed',marker='o')
plt.scatter(data_sort_rank['States/UTs'],data_sort_rank['DD_rank'],label='Drunken Driving',marker='+')
plt.scatter(data_sort_rank['States/UTs'],data_sort_rank['DW_rank'],label='Wrong Side Driving',marker='*')
plt.scatter(data_sort_rank['States/UTs'],data_sort_rank['JRL_rank'],label='Jumping Red Light',marker='^')
plt.scatter(data_sort_rank['States/UTs'],data_sort_rank['UM_rank'],label='Use of Mobile Phone',marker='.')
plt.xticks(rotation = 90)
plt.legend(title='Ranked based on lowest number of cases by',loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel('Number')
plt.grid()
plt.title('State-wise ranking based on cause of accident (lowest case ranked high)')
plt.show()
```

The above scatter plot shows that certain states or UTs are ranked high in all cases and some others ranked low in all cases. This can be a reflection of the stringent and lenient following of traffic rules in those states. Ranking of overspeeding and driving on the wrong side goes in a similar trend in most cases.

**Box Plot**

Dataset used is **'Sub Divisional Monthly Rainfall from 1901 to 2017'** from 'https://data.gov.in/resource/sub-divisional-monthly-rainfall-1901-2017'

```python
rain_data = pd.read_csv('Sub_Division_IMD_2017 - Sub_Division_IMD_2017.csv')
SI = ['Coastal Andhra Pradesh','Coastal Karnataka','North Interior Karnataka','South Interior Karnataka','Kerala','Tamil Nadu']
rain_data_SI = rain_data[rain_data['SUBDIVISION'].isin(SI)]

plt.figure(figsize=(10,8),dpi=100)
sns.boxplot(x='SUBDIVISION',y='JJAS',data=rain_data_SI,showmeans=True, hue = 'SUBDIVISION')
# plt.boxplot([1,2,3,4,5])
plt.xticks(rotation=90)
plt.ylabel('Rainfall in mm')
plt.xlabel('')
plt.title('Rainfall in South Indian subdivision for June to Sept. from 1901 to 2017')
plt.show()
```

The boxplot shows the rainfall for the period from June to September for South-Indian subdivisions. Coastal Karnataka receives the highest during the period followed by Kerala. The horizontal line represents the median and the green triangle represents the mean in the figure. Tamil Nadu receives the lowest rainfall among the subdivisions for the period.
